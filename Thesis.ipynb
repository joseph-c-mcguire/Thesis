## Function Definitions

````python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy
import datetime

def import_cases():
  !pip3 install sodapy
  import pandas as pd
  from sodapy import Socrata

  # Unauthenticated client only works with public data sets. Note 'None'
  # in place of application token, and no username or password:
  client = Socrata("data.cdc.gov", None)

  # Example authenticated client (needed for non-public datasets):
  # client = Socrata(data.cdc.gov,
  #                  MyAppToken,
  #                  userame="user@example.com",
  #                  password="AFakePassword")

  # First 2000 results, returned as JSON from API / converted to Python list of
  # dictionaries by sodapy.
  results = client.get("9mfq-cb36", limit = 100000)

  # Convert to pandas DataFrame
  results_df = pd.DataFrame.from_records(results)
  return results_df

  #!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

def import_vac():
  import pandas as pd
  from sodapy import Socrata

  # Unauthenticated client only works with public data sets. Note 'None'
  # in place of application token, and no username or password:
  client = Socrata("data.cdc.gov", None)

  # Example authenticated client (needed for non-public datasets):
  # client = Socrata(data.cdc.gov,
  #                  MyAppToken,
  #                  userame="user@example.com",
  #                  password="AFakePassword")

  # First 2000 results, returned as JSON from API / converted to Python list of
  # dictionaries by sodapy.
  results = client.get("rh2h-3yt2", limit=200000000)

  # Convert to pandas DataFrame
  vac_df = pd.DataFrame.from_records(results)
  return vac_df

def check_integrity(df):
  print(df.info())
  print(df.head())
  print(df.describe())
  for i in df.columns:
    print("Column: '" + i + "'")
    print(df[i].unique().size)
  for i in df.columns:
    print("Column NaN value count: '" + i + "'")
    print(str(round(df[i].isna().sum() / df.shape[0] * 100,2)) + '%')

def clean_data(df, date):
  import pandas as pd
  print(date)
  df_cleaned = df[['submission_date', 'state', 'tot_cases', 'new_case', 'tot_death', 'new_death','created_at']]
  df_cleaned['submission_date'] = pd.to_datetime(df_cleaned['submission_date'])
  df_cleaned['created_at'] = pd.to_datetime(df_cleaned['created_at'])
  df_cleaned[['tot_cases', 'new_case', 'tot_death', 'new_death']] = df_cleaned[['tot_cases', 'new_case', 'tot_death', 'new_death']].astype(float)

  def abbreviation_correction(row):
    if row['state'] == 'RMI':
      return 'MH'
    if row['state'] == 'NYC':
      return 'NY'
    if row['state'] == 'FSM':
      return 'FM'
    else:
      return row['state']

  df_cleaned['state'] = df_cleaned[['state']].apply(lambda row : abbreviation_correction(row) , axis = 1)
  aggregrate_function = {'tot_cases' : 'sum', 'tot_cases' : 'sum', 'new_case' : 'sum',
                        'tot_death' : 'sum', 'new_death' : 'sum', 'created_at' : 'first'}
  df_aggregrate = df_cleaned.groupby(['submission_date', 'state']).aggregate(aggregrate_function)
  df_cleaned = df_aggregrate.reset_index()
  df_cleaned = df_cleaned[df_cleaned.submission_date <= date]
  return df_cleaned

def normalize_by_pop(df):
  import pandas as pd
  import datetime

  df_pop = pd.read_csv(r'https://raw.githubusercontent.com/joseph-c-mcguire/Thesis/main/Paper/Pop_State2019Ages60andOlderPopulationEstimates.csv', usecols = [i for i in range(3)])
  df_pop_adjusted = df[df['state'].isin(df_pop.Abbreviations.unique())]
  df_pop['population'] = df_pop['Population Estimates - Total Population - Total Population - All Ages (Age 0 and Older)']
  df_pop_adjusted = df_pop_adjusted.merge(df_pop, left_on = 'state', right_on = 'Abbreviations',how = 'left')
  df_pop_adjusted.drop(['Abbreviations', 'Population Estimates - Total Population - Total Population - All Ages (Age 0 and Older)', 'Geography'], axis = 1, inplace = True)
  df = df_pop_adjusted
  df.population = df.population.astype(float)
  df['tot_cases_per'] = df.apply(lambda row : row['tot_cases'] / row['population'], axis = 1)
  df['new_case_per'] = df.apply(lambda row : row['new_case'] / row['population'], axis = 1)
  df['tot_death_per'] = df.apply(lambda row : row['tot_death'] / row['population'], axis = 1)
  df['new_death_per'] = df.apply(lambda row : row['new_death'] / row['population'], axis = 1)
  df['week_day'] = df.apply(lambda row : datetime.datetime.weekday(row['submission_date']) , axis = 1)

  return df

def make_cleaned_df(date,pop_adjusted = True):
  raw = import_cases()
  clean = clean_data(raw, date)
  if pop_adjusted:
    clean = normalize_by_pop(clean)
  return clean

def make_US_df(date):
  import pandas as pd
  import datetime
  raw_data = clean_data(import_cases(), date)
  df_cleaned = normalize_by_pop(raw_data)
  df_US = df_cleaned.groupby('submission_date').sum()
  df_US['tot_cases_per'] = df_US.apply(lambda row : row['tot_cases'] / row['population'], axis = 1)
  df_US['new_case_per'] = df_US.apply(lambda row : row['new_case'] / row['population'], axis = 1)
  df_US['tot_death_per'] = df_US.apply(lambda row : row['tot_death'] / row['population'], axis = 1)
  df_US['new_death_per'] = df_US.apply(lambda row : row['new_death'] / row['population'], axis = 1)
  df_US = df_US.reset_index()
  df_US['week_day'] = df_US.apply(lambda row : datetime.datetime.weekday(row['submission_date']) , axis = 1)
  df_US = df_US[df_US.submission_date <= date]
  return df_US
````

## Interpolation

````python
def transform(data):
  from numpy import log
  return log(data + 2)
def inverse_transform(data):
  from numpy import exp
  return exp(data) - 2
def check_stationary(data, alpha = 0.1):
  import statsmodels.api as ts
  # kpss_t = ts.tsa.kpss(data, 'ct')[1]
  kpss = (ts.tsa.kpss(data,'c')[1] <= alpha)
  # adfuller = (ts.tsa.kpss(data)[1] <= alpha)
  if kpss:
    return 0
  else:
    diff = data
    j = 0
    while not(kpss):
      diff = diff.diff().dropna()
      kpss = (ts.tsa.kpss(diff,'c')[1] < alpha)
      # adfuller = (ts.tsa.adfuller(data)[1] <= alpha)
      print(diff,kpss,j)
      j += 1
    return j
def impute_time_series(original, return_model_info = False):
  import statsmodels.api as ts
  import numpy as np 
  import matplotlib.pyplot as plt
  import pandas as pd

  # cleaned_df = make_cleaned_df()

  df = original
  states = df[df['new_case'] < 0].state.unique()
  model_info = pd.DataFrame({'state' : [], 'window' : [], 'order' : [], 'BIC' : []})
  windows = [30]
  predicted = np.array([])
  for i in states:
    state_df = df[df['state'] == i]
    index = state_df[state_df.new_case < 0].new_case.index.values
    state_df = state_df.reset_index()
    state_df.new_case = state_df.new_case.apply(lambda column : max(0, column))
    state_index = state_df[state_df['index'].isin(index)].index.values
    # for l in range(len(state_index)):
    #   temp = state_index[l]
    #   state_df.iloc[temp, state_df.columns.get_loc('new_case')] = (state_df.new_case.iloc[temp - 1] + state_df.new_case.iloc[temp + 1])/2
    data = state_df.new_case
    data = transform(data)
    # Baseline for comparison
    # mean = data.mean()
    for j in range(0,state_index.size):
      state_id = state_index[j]
      quality = np.zeros(len(windows))
      for k in range(len(windows)):
        window_length = windows[k]
        N_low = max(0,state_id - window_length)
        N_upp = min(state_df.shape[0], state_id + window_length)
        subset = data.iloc[N_low:state_id]
        difference = check_stationary(subset)
        arma_order = ts.tsa.arma_order_select_ic(subset).bic_min_order
        model = ts.tsa.ARIMA(data, (arma_order[0], difference, arma_order[1])).fit()
        quality[k] = model.bic
      ideal_window = np.argmin(quality)
      k = windows[ideal_window]
      N_low = max(0,state_id - k)
      N_upp = min(state_df.shape[0], state_id + k)
      subset = data.iloc[N_low:state_id]
      difference = check_stationary(subset)
      arma_order = ts.tsa.arma_order_select_ic(subset).bic_min_order
      model = ts.tsa.ARIMA(data, (arma_order[0], difference, arma_order[1])).fit()
      predictions = model.predict(state_id,state_id)
      predictions = inverse_transform(predictions)
      df.loc[index[j], 'new_case'] = predictions[state_index[j]]
      temp = pd.DataFrame({'state' : [i], 'window' : [k] ,'order' : [(arma_order[0], difference, arma_order[1])], 'BIC' : [model.bic]})
      model_info = model_info.append(temp)
  df.loc[:,'new_case_per'] = df.apply(lambda row : row['new_case'] / row['population'], axis = 1)
  if return_model_info:
    return (df, model_info)
  return df    
````

## Graph Building

Using state data on a daily basis, we'll construct our graph by calculating the correlation between states for each day in 2 differing manners, one in the way of the literature to date and the other a novel approach developed by myself:
<!-- 
We need to ensure that the data is stationary, which we do with the KPSS p-test. -->
- The Pearson r coefficient between $X_{i,t}, X_{j,t}$ for $t = 0,1,\ldots,14$ (literature)
- Cross correlation function calculates the Pearson $R$-score between two time series one with a delay, $X_t$ and $Y_{t - \tau}$ (me)

The first of these imposes that the graph must be symmetric in nature, while the second does not, allowing for directed behavior of pandemic spread.

### Function Definitions

````python
import numpy as np
import statsmodels.tsa.stattools as stattools
import datetime

def ccf_matrix(df, columns, window = datetime.datetime.now() - datetime.timedelta(days = 15)):
  state_names = df.state.unique()
  df = df[df['submission_date'] >= window]
  N = state_names.size
  T = df.submission_date.unique().size
  diff = np.zeros(len(columns))
  # category_names = ["new_case_per", "tot_cases_per", "new_death_per", "tot_death_per"]
  # new_case = np.ones(shape = (N,N,T))
  # tot_case = np.ones(shape = (N,N,T))
  # new_death = np.ones(shape = (N,N,T))
  # tot_death = np.ones(shape = (N,N,T))
  table = np.ones(shape = (len(columns), N,N,T))
  # print(len(table))
  # tables = np.array([new_case, tot_case, new_death, tot_death])
  for k in range(0,len(table)):
    column_name = columns[k]
    kpss_test = stattools.kpss(df[column_name])[0]
    while not(kpss_test < 0.1):
      diff[k] += 1
      print('Warning: ' + column_name + ' is not stationary! Operation we\'ll be preformed on differenced version')
      df[column_name + '_diff_' + str(diff[k])] = df[column_name].diff()
      kpss_test = stattools.kpss(df[column_name + '_diff_' + str(diff[k])])[0]
    for i in range(0,N):
      for j in range(0,N):
        df_i = df[df['state'] == state_names[i]]
        df_j = df[df['state'] == state_names[j]]
        table[k,i,j,:] =  stattools.ccf(df_i[column_name], df_j[column_name], unbiased = False)
    
  return (state_names, table, diff)
````

````python
def build_adjacency(ccf_table,t = 0,self_loops = False, 
                    weighted = False,
                    r = 0.5):
  if ccf_table.shape[0] != ccf_table.shape[1]:
    return 'Error'
  N = ccf_table.shape[0]
  # T = ccf_table.shape[2]
  A = np.zeros(shape = (N,N))
  for i in range(0,N):
    for j in range(0,N):
      if (np.abs(ccf_table[i,j,t]) >= r):
        if weighted:
          A[i,j] = ccf_table[i,j,t]
        else:
          A[i,j] = 1
      else: 
        A[i,j] = 0
  # if ~self_loops:
    # A = A - np.diag(np.ones(N))
  return A
````

Cross-Correlation function between the states is computed below, this is:
$$p_{XY}(\tau) = \dfrac{\sum\limits_{i = 0}^{N - 1} (X_i - \bar{X})(Y_{i - \tau} - \bar{Y})}{\sqrt{\sum\limits_{i = 0}^{N - 1} (X_i - \bar{X})^2} \sqrt{\sum\limits_{i = 0}^{N-1}(Y_i - \bar{Y})^2}} $$

This is the standardized correlation between the two time series $X$ and $Y$ with a lag of $\tau$ in $Y$.
So imporantly note that $p_{XY}(\tau) \neq p_{YX}(\tau)$.

This is a relevant statistic to us in the sense that it measures the correlation between time series $X_t$ and the lagged time series $Y_{t - \tau}$; that is, what is the response of $X$ based on a lag ($\tau$) of $Y$?
